## CLASS / OBJECT

```python
#Write a class with a perimeter and area function.
class circle:
 import math
 k=math.pi
 def perimeter(r):
   return 2*math.pi*r
 
 def area(r):
   return math.pi*r*r
  
print(circle.perimeter(10))
print(circle.area(10))
print(circle.k)
```
